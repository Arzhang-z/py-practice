- ContactBook:
- Methods:
    - add_contact (create)
    - view_contact (read)
    - update_contact (update)
    - delete_contact (delete)

in each database we need some info to be unique and keep it uncanged in order to keep track of each data inside the db while changing some part of it, but id/email/name(here we kept it as a unique and unchangable part) must not change

```python
from collections import defaultdict


class ContactBook:
    def __init__(self):
        #self.contacts = {}
        self.contacts: defaultdict[str,dict[str,str]]  = defaultdict(dict)

    def add_contact(self, name: str, phone: str, email: str = ""):
        if name in self.contacts:
            print("Contact Already Exists!")
            return

        #self.contacts[name] = {}
        self.contacts[name]["phone"] = phone
        if email:
            self.contacts[name]["email"] = email

    def view_contacts(self):
        for name, info in self.contacts.items():
            print(f"Name: {name}")
            print(f"phone: {info['phone']}")
            if info.get("email") is not None:
                print(f"Email: {info["email"]}")
            print("_" * 50)

    def delete_contact(self, name: str):
        if name in self.contacts:
            del self.contacts[name]
            print("Contact deleted successfully")
        else:
            print("Contact not found")

    def update_contact(self, name: str, phone: str = None, email: str = None):
        if name in self.contacts:
            if phone:
                self.contacts[name]["phone"] = phone
            if email:
                self.contacts[name]["eamil"] = email

            print("Contact updated successfully!")
            return
        
        print("Contact not found")
```

```python
book = ContactBook()


while True:
    print("\n\nWelcome to contact book application!")
    print("1. Add Contact")
    print("2. Edit Contact")
    print("3. View Contacts")
    print("4. Delete Contact")
    print("5. Quit")

    user_choice = input("Please choose an option: ")

    if user_choice == "5":
        break

    elif user_choice == "1":
        name = input("\nEnter contact name: ")
        phone = input("Enter contact phone: ")
        email = input("Enter contact email: ")

        book.add_contact(name, phone, email)

    elif user_choice == "2":
        name = input("\nEnter contact name: ")
        user_update_choice = input("Select which Item you want to update? \n1. phone \n2. email \n3. both")
        
        if user_update_choice == "1": 
            phone = input("Enter contact phone: ")
            book.update_contact(name, phone = phone)

        elif user_update_choice == "2":
            email = input("Enter contact email: ")
            book.update_contact(name, email = email)
        
        elif user_update_choice == "3":
            phone = input("Enter contact phone: ")
            email = input("Enter contact email: ")
            book.update_contact(name, phone, email)

    elif user_choice == "3":
        print("\nList of Contacts: ")
        book.view_contacts()
    
    elif user_choice == "4":
        name = input("\nEnter contact name: ")
        book.delete_contact(name)
```
