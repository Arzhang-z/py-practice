```python
import requests
```

```python
def check_site_availibilty(url_list):
    for url in url_list:
        try:
            headers = {
                'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; like Gecko) Chrome/58.0.3029.110 Safari/537.3'
            }
            response = requests.get(url,headers=headers, timeout=5)
            if response == 200:
                return f'{url} is up!'
            return f'{url} is down!'

        except Exception as e:
            return f'failed to reach {url}, Error: {e}'
```

```python
websites_to_check = [
    "https://www.google.com",
    "https://www.github.com",
    "https://www.python.org",
    "https://www.somenonexistingwebsite.io"
] *10
```

```python
check_site_availibilty(websites_to_check)
```
