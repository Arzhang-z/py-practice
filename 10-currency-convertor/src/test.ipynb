```python
def get_exchange_rate(base_currency: str = "CAD", target_currency: str = "USD") -> float:
    url = f'https://v6.exchangerate-api.com/v6/a3e11f53a075bc2556feab1f/latest/{base_currency}'
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()['conversion_rates'][target_currency]
    elif response.status_code/100 >= 5:
        print('Try Again Later, Server Error Occured')
        return 0
    else:
        print("Client Error!")
        return 0
```

```python
from functools import cache, lru_cache
```

@lru_cache(maxsize=None) == @cahhe

```python
@lru_cache(maxsize=3) # so in 1001 removes first
def fibonacci(n: int) -> int:
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
fibonacci(1000)
```

```python
fibonacci.cache_info()
```

```python
fibonacci.cache_clear()
```

```python
from cachetools import cached, TTLCache
```

```python
ttl_chache = TTLCache(maxsize=100, ttl=10)
```

```python
import time

@cached(ttl_chache)
def get_data(key):
    time.sleep(3)
    return f'DAta for {key}'
```
