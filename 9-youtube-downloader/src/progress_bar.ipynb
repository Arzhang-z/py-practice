```python
names = [
    "A",
    "B",
    "C",
    "D",
    "E",
    "F",
    "G",
    "H",
]
```

```python
import time
from tqdm import tqdm

pbar = tqdm(names)
for name in pbar:
    pbar.set_description(f"Processing {name}")
    pbar.update(1)
    #print(name)
    time.sleep(1)
```

```python
from tqdm import trange

for i in trange(100):
    time.sleep(0.1)
```
