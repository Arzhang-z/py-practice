```python
UNDER_20 = [
    "Zero", "One", "Two", "Three", "Four", "Five", "Six", "Seven", "Eight", "Nine",
    "Ten" , "Elevel", "Twelve", "Thirteen", "Fourteen", "Fifteen", "Sixteen",
    "Seventeen", "Eighteen", "Nineteen"
]
```

```python
TENS = [
    "", "", "Twenty", "Thirty", "Forty", "Fifty", "Sixty", "Seventy", "Eighty", "Ninety"
]
```

```python
ABOVE_100 = {
    100: "Hundred",
    1000: "Thousand",
    1000000: "Million",
    1000000000: "Billion",
    1000000000000: "Trillion"
}
```

```python
def check_valid():
    assert num_to_word(0) == "Zero"
    assert num_to_word(23) == "Twenty Three"
    assert num_to_word(148) == "One Hundred Forty Eight"
    assert num_to_word(26987) == "Twenty Six Thousand Nine Hundred Eighty Seven"
    assert num_to_word(100000000000) == "One Hundred Billion"
```

```python
def num_to_word(num: int):
    if num < 20:
        return UNDER_20[num]
    elif num < 100:
        return TENS[num // 10] + (" "+UNDER_20[num % 10] if num%10 != 0 else "")
    
    pivot = max([key for key in ABOVE_100.keys() if key <= num])
    p1 = num_to_word(num // pivot)
    p2 = ABOVE_100[pivot]

    # if num % pivot == 0:
    #     return f'{p1} {p2}'
    
    return f'{p1} {p2}{" " + num_to_word(num % pivot) if num % pivot != 0 else ""}'
```

```python
from cachetools import cached, TTLCache
ttl_cache = TTLCache(3000,30)

@cached(ttl_cache)
def factoriel(n):
    if n == 0:
        return 1
    else:
        print(f"calling factoriel({n-1}) from factoriel({n})")
        return n * factoriel(n-1)
```

```python
factoriel(5)
```

```python
from cachetools import cached, TTLCache

cache = TTLCache(4,3)
@cached(cache)
def fib(n):
    # base case: if nis 0 or 1, returns the same number
    if n <= 0:
        return 1
    elif n == 1 :
        return 1
    # recursie case: F(n) = F(n-1) + F(n-2)
    else:
        return fib(n-1) + fib(n-2)
```

```python
fib(1000)
```

```python
def sum_list(numbers: list[int]) -> int:
    if numbers == []:
        return 0
    else:
        head, *tail = numbers
        return head + sum_list(tail)
```

```python
def sum_of_natural_numbers(n: int):
    if n == 1:
        return 1
    return n + sum_of_natural_numbers(n-1)
```
