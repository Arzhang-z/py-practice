```python
while True:
    user_guest = input("Guess a number between 1 and 100: ")
    if user_guest == "q":
        print("Thank you for playing. Goodbye")
        break
    
    elif not user_guest.isdigit():
        print("Invalid input. Please try again. ")
        continue
    
    user_guest = int(user_guest)
    if user_guest > 100 | user_guest <= 0:
        print("Your guess is out of range, Please try again later, Your guess must be between 1 and 100.")
        continue

    print(user_guest)
```
